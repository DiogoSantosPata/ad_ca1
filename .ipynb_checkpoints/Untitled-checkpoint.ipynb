```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import scipy.io as io
from scipy.signal import medfilt
from scipy import stats

import os
```

```python
m12_c1 = io.loadmat('../data/Mouse12_chan123/Mouse12-chan1.mat' )['dat'][0]
m12_c2 = io.loadmat('../data/Mouse12_chan123/Mouse12-chan2.mat' )['dat'][0]
m12_c3 = io.loadmat('../data/Mouse12_chan123/Mouse12-chan3.mat' )['dat'][0]
angles = np.loadtxt('../data/AngFiles/Mouse12-120806.txt')
```

```python
length_factor =  m12_c1.size/angles.size
n_angle = np.repeat( angles , length_factor )
n_angle = np.degrees(n_angle)
n_angle[n_angle<0.0] += 360.
```

```python
tam = 1500000
tamanho = tam+10000

plt.plot( medfilt( m12_c1[tam:tamanho] , kernel_size=101 ) )
plt.plot( medfilt( m12_c2[tam:tamanho] , kernel_size=101 ) )
plt.plot( medfilt( m12_c3[tam:tamanho] , kernel_size=101 ) )

plt.show()
```

```python
receptive_angle_mean, receptive_angle_std = [],[]

for aa in range(0,360,18):

    indx = np.where(np.logical_and( n_angle>aa ,  n_angle<aa+1 ))[0]
    mmm = np.median( m12_c2[indx] )
    sss = stats.sem( m12_c2[indx] )
    
    receptive_angle_mean.append(mmm)
    receptive_angle_std.append(sss)
```

```python
plt.errorbar( np.arange(len(receptive_angle_mean)),  receptive_angle_mean, yerr=sss )

plt.xticks( np.arange(20) , np.arange(0,361,20) )

plt.show()
```
