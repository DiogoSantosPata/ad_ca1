```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import scipy.io as io
from scipy.signal import medfilt
from scipy import stats
import os
```

```python
TH1 = io.loadmat('../data/channels_Mouse12_session_0807/Mouse12-chan1.mat' )['dat'].T[0]
HPC1 = io.loadmat('../data/channels_Mouse12_session_0807/Mouse12-chan71.mat' )['dat'][0]

TH2 = io.loadmat('../data/channels_Mouse12_session_0807/Mouse12-chan2.mat' )['dat'][0]
HPC2 = io.loadmat('../data/channels_Mouse12_session_0807/Mouse12-chan73.mat' )['dat'][0]
```

```python
s_rate = 1250
awake_window = [ 5698.9 * s_rate , 7778.8 * s_rate ]
rem_window   = [ 2503.2 * s_rate , 2579   * s_rate ]
sws_window   = [ 1639.3 * s_rate , 1929.6 * s_rate ] 
```

```python
tt_data_awake = TH1[awake_window[0]:awake_window[1]]
hh_data_awake = HPC1[awake_window[0]:awake_window[1]]

tt_data_rem = TH1[rem_window[0]:rem_window[1]]
hh_data_rem = HPC1[rem_window[0]:rem_window[1]]

tt_data_sws = TH1[sws_window[0]:sws_window[1]]
hh_data_sws = HPC1[sws_window[0]:sws_window[1]]
```

```python
lfp_sr = 1250.
video_sr = 39.06
sr_diff = lfp_sr / video_sr 
```

```python
events = np.loadtxt('lmin_indx_valid_Mouse12_120807.csv', delimiter=',')   # events for movement local minima
events = events * sr_diff
events = events.astype(int)

angles = np.loadtxt('lmin_indx_valid_angle.csv', delimiter=';')
angles = angles * sr_diff
angles = angles.astype(int)
```

# Motion/angle correlation

```python
s_wind_lenght =  1250*.2

corr_bef_total, corr_aft_total, corr_rand_total = [],[],[]
for ii in range(len(events)):
    s_wind  = events[ii]
    s_rand_wind = np.random.randint(events[1], events[-2])

    corr_bef  =  stats.pearsonr( tt_data_awake[s_wind-s_wind_lenght:s_wind] , hh_data_awake[s_wind-s_wind_lenght:s_wind] )
    corr_aft  =  stats.pearsonr( tt_data_awake[s_wind:s_wind+s_wind_lenght] , hh_data_awake[s_wind:s_wind+s_wind_lenght] )
    corr_rand =  stats.pearsonr( tt_data_awake[s_rand_wind:s_rand_wind+s_wind_lenght] , hh_data_awake[s_rand_wind:s_rand_wind+s_wind_lenght] ) 

    corr_bef_total.append(  corr_bef[0] )
    corr_aft_total.append(  corr_aft[0] )
    corr_rand_total.append( corr_rand[0] )
    
    
ang_corr_bef_total, ang_corr_aft_total, ang_corr_rand_total = [],[],[]
for ii in range(len(angles)):
    s_wind  = angles[ii]
    s_rand_wind = np.random.randint(angles[1], angles[-2])

    corr_bef  =  stats.pearsonr( tt_data_awake[s_wind-s_wind_lenght:s_wind] , hh_data_awake[s_wind-s_wind_lenght:s_wind] )
    corr_aft  =  stats.pearsonr( tt_data_awake[s_wind:s_wind+s_wind_lenght] , hh_data_awake[s_wind:s_wind+s_wind_lenght] )
    corr_rand =  stats.pearsonr( tt_data_awake[s_rand_wind:s_rand_wind+s_wind_lenght] , hh_data_awake[s_rand_wind:s_rand_wind+s_wind_lenght] ) 

    ang_corr_bef_total.append(  corr_bef[0] )
    ang_corr_aft_total.append(  corr_aft[0] )
    ang_corr_rand_total.append( corr_rand[0] )
```

```python
plt.figure(figsize=(10,4))
plt.suptitle('AD-CA1 correlation is motion independent', fontsize=18, y=1.05)

plt.subplot(1,2,1)
plt.boxplot([corr_bef_total, corr_aft_total, corr_rand_total], sym='')
plt.xticks([1,2,3],['bef_mov','aft_mov','rand'])
plt.ylim(-1.,.2)

plt.subplot(1,2,2)
plt.boxplot([ang_corr_bef_total, ang_corr_aft_total, ang_corr_rand_total], sym='')
plt.xticks([1,2,3],['bef_ang','aft_ang','rand'])
plt.ylim(-1.,.2)
;
```
